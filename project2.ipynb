## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
import requests
import pandas as pd
from urllib.parse import urlparse
```

```python
#save and print the text content of a page with all tags removed
# URL of the Wikipedia page
url = "https://en.wikipedia.org/wiki/Alpine_marmot"

# Download the page content
response = requests.get(url)
soup = bs4.BeautifulSoup(response.text, 'html.parser')

page_text = soup.get_text(separator="\n")
with open("page_text.txt", "w", encoding="utf-8") as file:
    file.write(page_text)
print("Full text content of the page (tags removed) saved in 'page_text.txt':\n")
print(page_text)
```

```python
#download an image with beautifulsoup and save it in this repository
img_tag = soup.find('img')
if img_tag:
    img_url = img_tag.get('src')
    # Handle relative URL cases
    if img_url.startswith("//"):
        img_url = "https:" + img_url
    elif img_url.startswith("/"):
        img_url = "https://en.wikipedia.org" + img_url
 
    # Download the image
    img_response = requests.get(img_url)
    if img_response.status_code == 200:
        with open("downloaded_image.jpg", "wb") as img_file:
            img_file.write(img_response.content)
        print("\nImage downloaded and saved as 'downloaded_image.jpg'.")
    else:
        print("\nFailed to download the image. Status code:", img_response.status_code)
else:
    print("\nNo image found on the page.")
```

```python
#find all the links in a page with beautifulsoup
#print the first 100 characters of ten of these links
all_links = soup.find_all('a')
print("\nFirst 10 links (displaying the first 100 characters of each link's HTML):")
for i, link in enumerate(all_links[:10], start=1):
    link_html = str(link)
    print(f"\nLink {i}:")
    print(link_html[:100])
```

```python
def download_script_text(url):
    try:
        # Check if the URL has a valid scheme
        parsed = urlparse(url)
        if not parsed.scheme:
            print(f"Invalid URL {url!r}: No scheme supplied. Skipping.")
            return ""
        
        # Skip URLs pointing to PDF files (bs4 cannot parse PDF content)
        if url.lower().endswith('.pdf'):
            print(f"Skipping PDF URL: {url!r}")
            return ""
        
        # Download the page content using requests
        response = requests.get(url, timeout=10)
        if response.status_code != 200:
            print(f"Warning: Received status code {response.status_code} for URL: {url!r}")
            return ""
        
        # Parse the HTML with BeautifulSoup and remove all tags
        soup = bs4.BeautifulSoup(response.text, 'html.parser')
        text = soup.get_text(separator=" ", strip=True)
        
        # Return the first 1000 characters of the text
        return text[:1000]
    
    except Exception as e:
        print(f"Error processing URL: {url!r}\n{e}")
        return ""
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"

# Using the links in the "link" column, download the first 1000 characters of each script
scripts['script_text'] = scripts['link'].apply(download_script_text)

# Display a preview of the updated DataFrame
print("\nUpdated DataFrame with script_text column:")
print(scripts.head())

# Save this new DataFrame as "pudding_texts.csv"
scripts.to_csv("pudding_texts.csv", index=False)
print("\nNew DataFrame saved as 'pudding_texts.csv'")
```

```python
#create a dataset of the movies in theaters now. Include metadata fields you are interested in. 
import requests
import pandas as pd
import os

# TMDB API configuration
API_KEY = "cf5aed38ff80cbe45b0f891a68f7e504"  # Replace with your TMDB API key
base_url = "https://api.themoviedb.org/3"
now_playing_endpoint = f"{base_url}/movie/now_playing"

# Set the parameters for the API request
params = {
    "api_key": API_KEY,
    "language": "en-US",
    "page": 1,     
}

response = requests.get(now_playing_endpoint, params=params)
data = response.json()

movies = data.get("results", [])

movies_list = []
for movie in movies:
    movies_list.append({
        "id": movie.get("id"),
        "title": movie.get("title"),
        "overview": movie.get("overview"),
        "release_date": movie.get("release_date"),
        "vote_average": movie.get("vote_average"),
        "popularity": movie.get("popularity"),
        "poster_path": movie.get("poster_path")
    })

movies_df = pd.DataFrame(movies_list)
movies_df.to_csv("now_playing_movies.csv", index=False)
print("Dataset of movies in theaters saved as 'now_playing_movies.csv'.")

posters_dir = "posters"
if not os.path.exists(posters_dir):
    os.makedirs(posters_dir)

image_base_url = "https://image.tmdb.org/t/p/w500"
```

```python
#download the movie posters for 10 of these movies and save them to this repository
for index, row in movies_df.head(10).iterrows():
    poster_path = row["poster_path"]
    if poster_path:
        poster_url = f"{image_base_url}{poster_path}"
        try:
            img_response = requests.get(poster_url, timeout=10)
            if img_response.status_code == 200:
                # Create a safe filename using the movie id and title (replace spaces)
                filename = f"{posters_dir}/{row['id']}_{row['title'].replace(' ', '_')}.jpg"
                with open(filename, "wb") as img_file:
                    img_file.write(img_response.content)
                print(f"Downloaded poster for '{row['title']}' to {filename}")
            else:
                print(f"Failed to download poster for '{row['title']}'. Status code: {img_response.status_code}")
        except Exception as e:
            print(f"Error downloading poster for '{row['title']}': {e}")
    else:
        print(f"No poster available for movie: '{row['title']}'")
```
